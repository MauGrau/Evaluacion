```python
#Importamos librerías requeridas
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Convertimos el dataset en dataframe
Tokyo= pd.read_csv("Tokyo.csv")
Tokyo
```

```python
#a-Los registros de 5 “host name” diferentes
import pandas as pd

df = pd.read_csv("Tokyo.csv", low_memory=False)

filtro_a = df[df["host_name"].isin([
    "Yoshimi & Marek", "Kei", "Wakana", "Yume", "David"
])]

filtro_a.to_csv("filtro_a.csv", index=False)
filtro_a

#Variable filtrada: host_name
#Condición: Filtrar los registros de 5 host_name específicos: "Yoshimi & Marek", "Kei", "Wakana", "Yume", "David".
```

```python
#b-Los registros de los host que se hallan unido a Airbnb “despues del año 2020”

df["host_since"] = pd.to_datetime(df["host_since"], errors="coerce")

filtro_b = df[df["host_since"] > "2020-12-31"]

filtro_b.to_csv("filtro_b.csv", index=False)

filtro_b

# Variable filtrada: host_since
# Condición: Seleccionar registros de los anfitriones que se unieron después de 2020
```

```python
#c-Los registros de los host que responden “a más tardar en 1 día”

filtro_c = df[df["host_response_time"] == "within a day"]

filtro_c.to_csv("filtro_c.csv", index=False)

filtro_c

# Variable filtrada: host_response_time
# Condición: Seleccionar los registros de los anfitriones que respondieron "a más tardar en 1 día"
```

```python
#d-Los registros de las filas “Múltiplos de 200… (400, 600, 800, etc.)”
filtro_d = df[df.index % 200 == 0]

filtro_d.to_csv("filtro_d.csv", index=False)

filtro_d

# Variable filtrada: índice de filas
# Condición: Seleccionar las filas cuyo índice sea múltiplo de 200 (400, 600, 800, …).
```

```python
#e-Los registros de los tipos de cuarto: “Entire home” y “Private room”
filtro_e = df[df["room_type"].isin(["Entire home/apt", "Private room"])]
filtro_e.to_csv("filtro_e.csv", index=False)
filtro_e

# Variable filtrada: room_type
# Condición: registros cuyo tipo de cuarto sea "Entire home/apt" o "Private room".
```

```python
#f-Los registros que “a lo mucho cuenten con 4 camas”
filtro_f = df[df["beds"] <= 4]
filtro_f.to_csv("filtro_f.csv", index=False)
filtro_f

# Variable filtrada: beds
# Condición: registros donde el número de camas (beds) sea menor o igual a 4.
```

```python
#g-Los registros que “sean superhost”
# g - Los registros que sean superhost

filtro_g = df[df["host_is_superhost"] == "t"]

filtro_g.to_csv("filtro_g.csv", index=False)

filtro_g


# Variable filtrada: host_is_superhost
# Condición: Filtrar los registros donde el anfitrión es un Superhost (host_is_superhost = "t").
```

```python
#h-Los registros de las columnas “impares”
filtro_h = df.iloc[:, ::2]  # Selecciona todas las filas y columnas impares

filtro_h.to_csv("filtro_h.csv", index=False)

filtro_h

#Variable filtrada: el índice de las columnas (posición en el DataFrame).
#Condición aplicada: seleccionar únicamente las columnas en posiciones impares (1, 3, 5, 7, …).
```
